**Importando o pandas**

```python
import pandas as pd
```

**Criando um dataframe a partir de um dicionário**

```python
# dataframe = pd.DataFrame()
venda = {'data': ['15/02/2025', '16/02/2025'],
         'valor': [500, 300],
         'produto': ['caneta', 'papel'],
         'qtd': [50, 70],
        }
vendas_df = pd.DataFrame(venda)
```

**Visualização dos Dados**
 - print
 - display

```python
print(vendas_df)
```

```python
display(vendas_df)
```

**Importando arquivos e bases de dados** 

```python
vendas_df = pd.read_excel('Vendas.xlsx')
display(vendas_df)
```

**Resumos de Visualização de Dados simples e úteis**
- head
- shape
- describe

```python
display(vendas_df.head(10))
print(vendas_df.shape)
display(vendas_df.describe())
```

**Isolando colunas (e os pd.Series)**

```python
produtos = vendas_df[['Produto', 'ID Loja']]
display(produtos)
```

**Método .loc**
- Isolando uma linha
- Isolando linhas e colunas condicionadas
- Isolando 1 valor específico

```python
# Isolando linha
display(vendas_df.loc[1])
```

```python
display(vendas_df.loc[7:9])
```

```python
# Isolando linhas condicionadas
# vendas_norteshopping_df
display(vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping'])
```

```python
# Isolando colunas condicionadas
display(vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping', ['ID Loja', 'Produto', 'Quantidade']])
```

```python
# Isolando um valor específico
print(vendas_df.loc[789, 'Produto'])
```

**Adicionando Colunas**

```python
# a partir de uma coluna que existe
vendas_df['Comissão'] = vendas_df['Valor Final'] * 0.05
display(vendas_df)
```

```python
# coluna com valor padrão
vendas_df.loc[:,'Imposto'] = 0
display(vendas_df)
```

**Adicionando linhas**

```python
vendas_dez_df = pd.read_excel('Vendas - Dez.xlsx')
display(vendas_dez_df)
```

```python
vendas_df = pd.concat([vendas_df, vendas_dez_df])
display(vendas_df)
```

**Excluindo linhas e colunas**

```python
# Excluindo coluna
vendas_df = vendas_df.drop('Imposto', axis=1)
display(vendas_df)
```

```python
# Excluindo linha
vendas_df = vendas_df.drop(0, axis=0)
display(vendas_df)
```

**Valores Vazios**
- Deletando linhas/colunas vazias
- Deletando linhas que possuem valores vazios
- Preenchendo valores

```python
# Deletando linhas e colunas completamente vazias
#vendas_df = vendas_df.dropna(how= 'all', axis=1)

# Deletando linhas que possuem pelomenos 1 valor vazio
#vendas_df = vendas_df.dropna()

# Preenchendo valores vazios

#-Com o útlimo valor
#vendas_df = vendas_df.ffill()

#-Com a média da coluna
vendas_df['Comissão'] = vendas_df['Comissão'].fillna(vendas_df['Comissão'].mean())
display(vendas_df)
```

**Calcular Indicadores**
- Groupby
- Value Counts

```python
# Value Count
transacoes_loja = vendas_df['ID Loja'].value_counts()
display(transacoes_loja)
```

```python
# Groupby
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_produto)
```

**Mesclar 2 dataframes (procurando informações)**

```python
gerentes_df = pd.read_excel('Gerentes.xlsx')
display(gerentes_df)
```

```python
vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```
